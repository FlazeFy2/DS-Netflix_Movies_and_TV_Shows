```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("netflix_titles.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Total Series by Age Rating Category Comparison
df = pd.read_csv("netflix_titles.csv")

# Filter out invalid ratings (start with number and nan)
df = df.dropna(subset=['rating'])
df = df[~df['rating'].str.match(r'^\d')]

# Map ratings to categories
rating_groups = {
    "All Age": ['G', 'TV-G', 'TV-Y', 'TV-Y7', 'TV-Y7-FV'],
    "Under 17 Allowed": ['PG', 'PG-13', 'TV-PG', 'TV-14'],
    "Adults Only": ['R', 'NC-17', 'TV-MA', 'NR', 'UR']
}
df['rating_group'] = df['rating'].map(
    lambda x: next((key for key, values in rating_groups.items() if x in values), 'Unknown')
)

# Dataset
top_context = df['rating_group'].value_counts()

# Pie Chart
plt.figure(figsize=(8, 8))
top_context.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Total Series by Age Rating Category Comparison")
plt.ylabel("")
plt.show()
```
